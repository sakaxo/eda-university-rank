### Data cleaning

Inspecting both head and tail of the dataset, there some missing values in all the columns including the **Name of University** column.

#### Handling missing values in scores.

From **Teaching Score** through to **International Outlook Score**

1.  replace all missing scores with average score
2.  make any <font color= red>score greater than 100 equals 100 </font> and for <font color= red>those less than 0 be 0</font>.

```python
# get average scores
average_Teaching_Score = df['Teaching Score'].mean()
average_Research_Score = df['Research Score'].mean()
average_Citations_Score = df['Citations Score'].mean()
average_Income_Score = df['Industry Income Score'].mean()
average_Outlook_Score = df['International Outlook Score'].mean()
```

```python
# replace nan scores with average scores
df = df.copy() # just to avoid SettingWithCopyWarning
df['Teaching Score'].fillna(average_Teaching_Score, inplace=True)
df['Research Score'].fillna(average_Research_Score, inplace=True)
df['Citations Score'].fillna(average_Citations_Score, inplace=True)
df['Industry Income Score'].fillna(average_Income_Score, inplace=True)
df['International Outlook Score'].fillna(average_Outlook_Score, inplace=True)
```

Data from **Teaching Score** through to **International Outlook Score** is clean and good to use.

#### Handling missing values in **No of student per staff**.

1. fill all the NaN values with mode

```python
mode_student_per_staff = df['No of student per staff'].mode()[0]
df['No of student per staff'].fillna(mode_student_per_staff, inplace=True)
```

```python
df.tail()
```

#### Universities with unknown  locations
About  <font color =red>132</font> universities have NaN(empty) locations.
> Remove those by filtering

```python
# remove uni. with NaN location.
df = df[df.Location.notnull()]
```

```python
df
```

### OverAll Score cleaning

This colums has empty values and also data n wrong format. Eg. <font color =red>26.4-45.3</font>.

**Cleaning process**

1. finding average of minimum and maximum value of every wrong format
2. replace wrong format with calculated average
3. change dtype from Object to float
4. fill na values


**finding average**

```python
f = df[df['OverAll Score'].str.contains(r'\d+\.\d+–\d+\.\d+', na=False)] #get all wrong format values
```

```python
#split values and find average of min and max values

fs = f['OverAll Score'].str.split('–', expand=True) 

f_average = (pd.to_numeric(fs[0]) + pd.to_numeric(fs[1]))/2.
```

**replace wrong format with computed average**

```python
dfd = df.copy() # to avoid SettingWithCopyWarning

dfd.loc[f.index, 'OverAll Score'] = f_average
```

**Change dtype from object to float**

```python
print(f"Current dtype: {dfd['OverAll Score'].dtypes}")
```

```python
# change object type too float

dfd.loc[:,'OverAll Score'] = dfd['OverAll Score'].astype(float)
```

```python
print(f"dtype after change: {dfd['OverAll Score'].dtypes}")
```

**fill empty values with mean from this column**

```python
dfd['OverAll Score'].fillna(dfd['OverAll Score'].mean(), inplace=True)
```

### Handling (<font color=red>,</font>) in No of student

some values have commas(,)
1. remove commas(,)
2. change dtype to int
3. fill in missing values, if any


```python
# replace comma with nothing
dfd['No of student'] = dfd['No of student'].str.replace(pat=',', repl='')
```

```python
dfd.tail() #looking good
```

```python
# change dtype to int
dfd['No of student'] = dfd['No of student'].astype(int)
```

```python
# checking for empty values
dfd[dfd['No of student'].isna()]
```

```python
# huuuurrrrrrrrrrrrrrrrrrrry, there is no empty values
```

*Inserting new column to hold number of* **International Student**, *rather than the number in percentage*

```python
# insert column next to International Student
dfd.insert(6, "Number of International Student", 0)

```

### International Student
1. remove % symbol
2. convert type to float

#### Number of International Student

> **No of student** by **International Student** /100

```python
dfd['International Student'] = dfd['International Student'].str.replace(pat='%', repl='')
```

```python
# to numeric dtype
dfd['International Student'] = pd.to_numeric(dfd['International Student'])
```

```python
dfd['Number of International Student'] = dfd['No of student']*(dfd['International Student']/100)
```

```python
dfd.head()
```


### Adding two more columns to hold Number of Female and Male.
>The values will be Female:Male ratio converted to number

```python
female_male_ratio  = dfd['Female:Male Ratio'].str.split(':', expand= True)

female_ratio = pd.to_numeric(female_male_ratio[0])
male_ratio   = pd.to_numeric(female_male_ratio[1])
```

```python
# ratio to actual number
no_of_female = dfd['No of student']*(female_ratio/100)
no_of_male = dfd['No of student']*(male_ratio/100)
```

**add columns now**

```python
dfd.insert(8, "Female", no_of_female) # add number of female column
```

```python
dfd.insert(9, "Male", no_of_male) # add number of male column
```

```python
dfd.loc[10,:]
```

```python
# fill nan values with interpolate
dfd["Female"] = dfd["Female"].interpolate()
dfd["Male"] = dfd["Male"].interpolate()
```

```python
# round few columns to whole numbers
dfd["Female"] = dfd["Female"].round()
dfd["Male"] = dfd["Male"].round()
dfd["Number of International Student"] = dfd["Number of International Student"].round()
```

```python
dfd.tail() 
```

```python
# drop few columns 
dfd.drop(['International Student','Female:Male Ratio'], axis=1, inplace=True)

```

```python

dfd.drop_duplicates(subset=['Name of University'], inplace=True)
```
